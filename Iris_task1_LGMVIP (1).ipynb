## **construction of Scatter Plot**

---



```python
data_set["Color"] = data_set["Species"].map({"Iris-setosa": "red", "Iris-versicolor": "green", "Iris-virginica": "blue"})
```

```python
sns.FacetGrid(data_set, hue="Color", height=5, palette="Set2").map(plt.scatter, "SepalLengthCm", "PetalLengthCm").add_legend()
```

```python
sns.FacetGrid(data_set, hue="Color", height=5, palette="Set2").map(plt.scatter, "SepalWidthCm", "PetalWidthCm").add_legend()
```

```python
x = data_set[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]].values
y = data_set[["Species"]].values
Model = LogisticRegression()
Model.fit(x, y)
```

### `**Finding the Accuracy**

---






```python
Model.score(x,y).round(2)
```

## **summarize the fit of the model**

---



```python
Actual = y
predicted = Model.predict(x)
from sklearn import metrics
print (metrics.classification_report (Actual, predicted))
predicted = Model.predict(x)
from sklearn import metrics
print (metrics.classification_report (Actual, predicted))
```
